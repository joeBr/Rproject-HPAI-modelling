```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
poultry_outbreaks = pd.read_excel("data\poultry_outbreak_data.xlsx").iloc[::-1]
```

```python
mpl.rc('image', cmap='viridis')
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 40}

mpl.rc('font', **font)
```

```python
frequencies = poultry_outbreaks["ip"]
freq_series = pd.Series(frequencies)

x_labels = poultry_outbreaks["start_year"]
x_labels.iloc[6] = "2022/2023"+  u"\u2021"

# Plot the figure.
plt.figure(figsize=(30, 20))
ax = freq_series.plot(kind="bar",color = ["b","b","b","b","b","r","r"],width = 0.9)
ax.set_xlabel("HPAI Season")
ax.set_ylabel("Infected Premises")
ax.set_xticklabels(x_labels)
ax.set_ylim([0,180])


rects = ax.patches

# Make some labels.
labels = poultry_outbreaks["ip"]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(
        rect.get_x() + rect.get_width() / 2, height + 2, label, ha="center", va="bottom"
    )
    
plt.gcf().subplots_adjust(bottom=0.2)

plt.savefig("images/UK_poultry_outbreaks_bar.png",dpi=150)
```
